#Keeping only Selected Label and itemid by Merging Both DF

```python
labitems_data_demo=pd.merge(selected_lebel,labitems_data_demo,on='label',how='left')
labitems_data_demo=labitems_data_demo.dropna(subset=['itemid'])#Removed itemid's Nan row
labitems_data_demo
```

#First merge between Labevents data and labitems data

```python

first_merge=pd.merge(labevents_data_demo,labitems_data_demo,on='itemid',how='left')
first_merge_1=first_merge.dropna(subset=['label'])#Droping 'label' column's nan value only
first_merge_1

```

#Second Merge between First merged data and propensity data

```python

second_merge=pd.merge(first_merge_1,propensity_data_demo,on='subject_id',how='left')#Merging first_merge_1 dataframe with propensity_data using subject_id
second_merge_1=second_merge.dropna(subset=['admittime'])#Droping 'admittime' column's nan value only
second_merge_1['charttime'] =  pd.to_datetime(second_merge_1['charttime'], infer_datetime_format=True)#Converting this column into datetime
second_merge_1['storetime'] =  pd.to_datetime(second_merge_1['storetime'], infer_datetime_format=True)#Converting this column into datetime
second_merge_1['admittime'] =  pd.to_datetime(second_merge_1['admittime'], infer_datetime_format=True)#Converting this column into datetime
second_merge_1['dischtime'] =  pd.to_datetime(second_merge_1['dischtime'], infer_datetime_format=True)#Converting this column into datetime
base_data=second_merge_1.copy()#Storing data to a new dataframe
base_data
```

#This snippet is for makeing Timeperiod condition

```python
# base_data['dif']=base_data['admittime']-base_data['storetime']#Creating new column for giving Timeperiod conditions
# base_data['Timeperiod']=np.nan#Creating and assigning null value to Timeperiod column
# base_data.loc[(base_data['dif'] > timedelta(days =0)) & (base_data['dif'] <= timedelta(days = 180) ),['Timeperiod']]='T1'#Condition for T1
# base_data.loc[(base_data['dif'] > timedelta(days =180)) & (base_data['dif'] <= timedelta(days = 365) ),['Timeperiod']]='T2'#Condition for T2
# base_data.loc[(base_data['dif'] <= timedelta(days =0)) & (base_data['dif']>=timedelta(days = -2) ),['Timeperiod']]='D1'#Condition for D1
# final_base_data=base_data.dropna(subset=['Timeperiod'])#Droping 'Timeperiod' column's nan value only
# final_base_data
```

#This portion excludes Timeperiod 

```python
base_data['Timeperiod']=np.nan#Creating and assigning null value to Timeperiod column
base_data.loc[(base_data['charttime'] >= base_data['admittime']) & (base_data['charttime'] <= base_data['dischtime']),['Timeperiod']]='Valid' #valid when starttime is in between admittime and dischtime
final_base_data=base_data.dropna(subset=['Timeperiod'])#Droping 'Timeperiod' column's nan value only
final_base_data['cnt']=1
final_base_data
```

#Making Pivot Table

```python
table = pd.pivot_table(final_base_data, values='cnt', index=['subject_id','hadm_id','admittime','dischtime'],columns=['label'], aggfunc='sum')
table
```
